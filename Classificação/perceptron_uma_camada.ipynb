```
import tensorflow as tf
import numpy as np
```

```
#valores de entrada
entrada = np.array( [[0, 0],
                     [0, 1],
                     [1, 0],
                     [1, 1]])
entrada
```

```
#valores esperados na saída

saida_esperada = np.array([[0.0],[0.0],[0.0],[1.0]])
saida_esperada
```

```
pesos = tf.Variable(tf.zeros([2, 1], dtype = tf.float64))#tf.zeros cria uma matriz de zeros com 2 linhas e 1 coluna
```

```
def step(x):
  return tf.cast(tf.to_float(tf.math.greater_equal(x, 1)), tf.float64)
```

```
init = tf.global_variables_initializer()
```

```
camada_saida = tf.matmul(entrada, pesos)
```

```
camada_saida_ativacao = step(camada_saida)
```

```
erro = tf.subtract(saida_esperada, camada_saida_ativacao)
```

```
delta = tf.matmul(entrada, erro, transpose_a= True) #transpose_a transpõe a matriz x
```

```
treinamento = tf.assign(pesos, tf.add(pesos ,tf.multiply(delta, 0.1))) #criando novos pesos para a rede neural
```

```
with tf.Session() as sess:
  sess.run(init)
  #print(sess.run(pesos))
  #print(sess.run(camada_saida))
  #print('\n')
  #print(sess.run(camada_saida_ativacao))
  #print('\n')
  #print(sess.run(erro))
  print(sess.run(treinamento))
  
  epoca = 0
  
  for i in range(15):
    epoca += 1
    
    erro_total, _ = sess.run([erro, treinamento])
    erro_soma = tf.reduce_sum(erro_total)
    print("epoca: ", epoca, "Erro: ", sess.run(erro_soma))
    if erro_soma.eval() == 0:
      break
    pesos_final = sess.run(pesos)
```

```
#teste
camada_saida_teste = tf.matmul(entrada, pesos_final)
camada_saida_ativacao_teste = step(camada_saida_teste)
with tf.Session() as sess:
  sess.run(init)
  print(sess.run(camada_saida_ativacao_teste))
```
