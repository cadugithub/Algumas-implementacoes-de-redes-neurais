```
import tensorflow as tf
import numpy as np
```

```
#valores de entrada
entrada = np.array( [[0, 0], [0, 1], [1, 0], [1, 1]])
entrada
```

```
#valores esperados na sa√≠da

saida_esperada = np.array([[1],[0],[0],[1]])
saida_esperada
```

```
neuronio_entrada = 2
neuronio_oculta = 3
neuronio_saida = 1
```

```
#                                                [linha]           [coluna]
pesos = {'oculta': tf.Variable(tf.random_normal([neuronio_entrada, neuronio_oculta]), name= 'pesos_oculta'),
         'saida' : tf.Variable(tf.random_normal([neuronio_oculta, neuronio_saida]), name = 'pesos_saida')}
```

```
type(pesos['oculta'])
```

```
init = tf.global_variables_initializer()
```

```
bias = {'oculta': tf.Variable(tf.random_normal([neuronio_oculta]), name = 'bias_oculta'),
        'saida' : tf.Variable(tf.random_normal([neuronio_saida]), name = 'bias_saida')}
```

```
xph = tf.placeholder(tf.float32, [4, neuronio_entrada], name='xph')
yph = tf.placeholder(tf.float32, [4, neuronio_saida], name='yph')
```

```
camada_oculta = tf.add(tf.matmul(xph, pesos['oculta']), bias['oculta'])
camada_oculta_ativacao = tf.sigmoid(camada_oculta)
camada_saida = tf.add(tf.matmul(camada_oculta_ativacao, pesos['saida']), bias['saida'])
camada_saida_ativacao = tf.sigmoid(camada_saida)
```

```
erro = tf.losses.mean_squared_error(yph, camada_saida_ativacao) # os argumentos do tf.losses.mean_squared_error(respostasVerdadeiras, respotaDoAlgoritmo) 
otimizador = tf.train.GradientDescentOptimizer(learning_rate=0.3).minimize(erro)
```

```
with tf.Session() as sess:
  sess.run(init)
  #print(sess.run(pesos['oculta']))
  #print(sess.run(pesos['saida']))
  #print(sess.run(camada_oculta, feed_dict= {xph: entrada}))
  #print(sess.run(camada_oculta_ativacao, feed_dict= {xph: entrada}))
  #print(sess.run(camada_saida, feed_dict= {xph: entrada}))
  #print(sess.run(camada_saida_ativacao, feed_dict= {xph: entrada}))
  for epoca in range(10000):
    erro_medio = 0
    _, custo = sess.run([otimizador, erro], feed_dict={xph: entrada, yph:saida_esperada})
    if epoca % 200 == 0:
      erro_medio += custo/4
      print(erro_medio)
  peso_final, bias_final = sess.run([pesos, bias])
```
